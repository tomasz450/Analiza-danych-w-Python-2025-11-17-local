
__Wskazówka:__ Użyj funkcji _np.all()._

````python
A = np.array([[3, 2, 1, 4],
              [5, 2, 1, 6]])

B = np.array([[3, 2, 1, 4],
              [5, 2, 0, 6]])

C = np.array([[True, False, False],
              [True, True, True]])

D = np.array([0.1, 0.3])
````

````python
for name, array in zip(list('ABCD'), [A, B, C, D]):
    print(f"Wynik dla macierzy {name}: {np.all(array)}")
````
